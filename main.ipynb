```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=42)
```

```python
from sklearn.preprocessing import OneHotEncoder,OrdinalEncoder
```

```python
from sklearn.compose import ColumnTransformer
```

```python
df['cp'] = df['cp'].replace('non-anginal', 'non-anginal pain')
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, OrdinalEncoder

transformer = ColumnTransformer(
    transformers=[
        # Nominal
        ('nominal',
         OneHotEncoder(sparse_output=False, drop='first'),
         ['dataset', 'restecgFeatured']
        ),

        # Ordinal
        ('slope_ord',
         OrdinalEncoder(categories=[['upsloping', 'flat', 'downsloping']]),
         ['slopeFeatured']
        ),

        ('cp_ord',
         OrdinalEncoder(categories=[[
             'typical angina',
             'atypical angina',
             'non-anginal pain',
             'asymptomatic'
         ]]),
         ['cp']
        ),

        ('thal_ord',
         OrdinalEncoder(categories=[[
             'normal',
             'fixed defect',
             'reversable defect'
         ]]),
         ['thal1chFeatured']
        )
    ],
    remainder='passthrough'
)
```

```python
X_train_transformed = transformer.fit_transform(X_train)
X_test_transformed  = transformer.transform(X_test)
```

```python
X_train_transformed
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train_transformed)
X_test_scaled  = scaler.transform(X_test_transformed)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logistic=LogisticRegression(
    class_weight='balanced',
    random_state=42,
    max_iter=2000
)
```

```python
logistic.fit(X_train_scaled,y_train)
```

```python
y_pred_NL=logistic.predict(X_test_scaled)
```

```python
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
Cm=confusion_matrix(y_test,y_pred_NL)
print(Cm)
print(classification_report(y_test,y_pred_NL))
score=accuracy_score(y_test,y_pred_NL)
print(score)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
model=LogisticRegression()
penalty=['l1','l2','elasticnet']
C_values=[100,1.0,0.1,0.01]
solver=['saga','sag','newton-cg','liblinear','lbfgs']
```

```python
params=dict(penalty=penalty,C=C_values,solver=solver)
```

```python
from sklearn.model_selection import StratifiedKFold
cv1=StratifiedKFold()
```

```python
grid=GridSearchCV(estimator=model,param_grid=params,cv=cv1,scoring='accuracy',n_jobs=-1)
```

```python
grid.fit(X_train_scaled,y_train)
```

```python
y_pred_grid=grid.predict(X_test_scaled)
```

```python
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
Cm=confusion_matrix(y_test,y_pred_grid)
print(Cm)
print(classification_report(y_test,y_pred_grid))
score=accuracy_score(y_test,y_pred_grid)
print(score)
```

```python
grid.best_estimator_
```

```python
grid.best_params_
```

```python
grid.best_score_
```

```python
from sklearn.model_selection import RandomizedSearchCV
random=RandomizedSearchCV(estimator=model,param_distributions=params,scoring='accuracy',n_jobs=-1,cv=5)
```

```python
random.fit(X_train_scaled,y_train)
```

```python
y_pred_random=random.predict(X_test_scaled)
```

```python
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
Cm=confusion_matrix(y_test,y_pred_random)
print(Cm)
print(classification_report(y_test,y_pred_random))
score=accuracy_score(y_test,y_pred_random)
print(score)
```
